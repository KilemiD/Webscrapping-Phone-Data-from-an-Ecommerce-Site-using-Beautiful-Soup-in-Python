```python
#our job now is to find these together and make a dataframe
#let's repeat the steps above but assign this time
#filter by name
name=soup.find_all('h3',class_='name')

#filter by price
new_price=soup.find_all('div',class_='prc')

#filter by old price
old_price=soup.find_all('div',class_='old')
```

```python
#what is the length of these lists, will help in our data frame creation
print('Phone Name',len(name),'New Price',len(new_price),'Old Price',len(old_price))
```

```python
#see from above the dimension is different
#the output from name looks like this
name
```

```python
#we need to only get the name of the phone and drop all other things
name_1=soup.find('h3',class_='name')
name_1.text
```

```python
#creating a for loop to use .text in the list of phone names
#phone name
product_name=[]
for i in name:
    nam=i.text
    product_name.append(nam)
    
#new price
price_list=[]
for i in new_price:
    price=i.text
    price_list.append(price)
    
#old price
price_old=[]
for i in old_price:
    old=i.text
    price_old.append(old)
```

```python
#lets see one
product_name
```

```python
#let us remove the empty parts of the string list
product_name2=[]
for string in product_name:
    if (string !=""):
        product_name2.append(string)
len(product_name2)
```

```python
#removing empty strings in price
price_list2=[]
for string in price_list:
    if (string !=""):
        price_list2.append(string)
len(price_list2)
```

```python
#checking product name again, note I am using the new product name
product_name2
```

```python
#you can see the first name of each row in the array is the phone brand
#we want to get that first name as our product brand
brand=list(next(zip(*map(str.split,product_name2))))
brand
```

```python
#now our job is done, we have 4 arrays that we want to use to make
#a data frame
import pandas as pd
data=pd.DataFrame({'Phone Name':product_name2,
                   'New Price':price_list2,
                   'Old Price':price_old,
                   'Brand Name':brand
})
```

```python
#we can now write a csv file
data.to_csv("Jumia Phones Data.csv")
```
